# Exploring TEI Structure of JQA

Create a Network Graph of Parent-Child TEI relationships.

```python
# Import necessary libraries.
import re, glob, csv, sys, os, json
import pandas as pd
from lxml import etree

from json import JSONEncoder
import seaborn as sns
import matplotlib.pyplot as plt

import networkx as nx
from networkx.algorithms import community
from networkx.readwrite import json_graph

from sklearn import preprocessing

# Declare directory location to shorten filepaths later.
abs_dir = "/Users/quinn.wi/Documents/"

# Gather all .xml files using glob.
list_of_files = glob.glob(abs_dir + "Data/JQA/*/*.xml")
```

```python
%%time

tei = []

# Loop through each file within a directory.
for file in list_of_files:

    tree = etree.parse(file)
    
    for elem in tree.iter():
        try:
            child = re.sub('{.*}(.*)', '\\1', str(elem.tag))
            parent = re.sub('{.*}(.*)', '\\1', str(elem.getparent().tag))
            
            data = {
                'source': parent,
                'target': child
            }
        
            tei.append(data)
            
        except AttributeError:
            pass

# # Manually add 'root' for d3.stratify()
# tei.append({
#     'parent': '',
#     'node': 'TEI'
# })

tei = pd.DataFrame(tei)

tei = tei.groupby(['source', 'target']).size().reset_index()
tei.columns = ['source', 'target', 'weight']

# tei['id'] = tei.index

tei.head()
```

```python
tei.query('weight <= 10')
```

```python
%%time

# Filter
print (f'Dataframe before filtering: {tei.shape}')
tei = tei.query('weight >= 300')

print (f'Dataframe after filtering: {tei.shape}')

tei.to_csv(abs_dir + 'GitHub/dsg-mhs/d3/Network/TEI-Structure/jqa_tei-structure.csv',
           sep = ',', 
           index = False)
```

## Convert to Network Object

```python
%%time

print (f'Min Weight: {min(tei.weight)}\nMax Weight: {max(tei.weight)}\n')

# Scale weights
scaler = preprocessing.MinMaxScaler()
tei['weight'] = scaler.fit_transform(tei[['weight']])

# Graph distribution
fig, ax = plt.subplots()
sns.set(rc={'figure.figsize': (16, 6)})

plt.hist(tei['weight'])
plt.show()
```

```python
%%time

G = nx.from_pandas_edgelist( tei[['source', 'target', 'weight']],
                            'source', 'target', 'weight')

# Add nodes.
nodes = list( dict.fromkeys( tei['source'].values.tolist() + tei['target'].values.tolist() ))
nodes = pd.DataFrame( nodes, columns = ['source'])
G.add_nodes_from(nodes)

# Set degree attributes.
nx.set_node_attributes(G, dict(G.degree(G.nodes())), 'degree')

# Centrality measures.
betweenness_dict = nx.betweenness_centrality(G)
eigenvector_dict = nx.eigenvector_centrality(G)
degree_centrality_dict = nx.degree_centrality(G)

# Assign measures.
nx.set_node_attributes(G, betweenness_dict, 'betweenness')
nx.set_node_attributes(G, eigenvector_dict, 'eigenvector')
nx.set_node_attributes(G, degree_centrality_dict, 'degree_centrality')


# Community (modularity)
communities = community.greedy_modularity_communities(G)

modularity_dict = {} # Create a blank dictionary
for i,c in enumerate(communities): # Loop through the list of communities, keeping track of the number for the community
    for name in c: # Loop through each person in a community
        modularity_dict[name] = i # Create an entry in the dictionary for the person, where the value is which group they belong to.

# Now you can add modularity information like we did the other metrics
nx.set_node_attributes(G, modularity_dict, 'modularity')
```

```python
%%time

with open(abs_dir + 'GitHub/dsg-mhs/d3/Network/TEI-Structure/jqa_tei-structure.json', 'w') as outfile:
    
    outfile.write( json.dumps(json_graph.node_link_data(G)))
```
