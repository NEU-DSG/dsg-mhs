# Automated Cleanup Processes for WETVAC -- Regex

```python
import re, glob, csv, textract
import pandas as pd

abs_dir = "/Users/quinn.wi/Documents/"
list_of_files = glob.glob(abs_dir + "Data/PSC/Taney/RBT_WET_08-2021/*.docx")
```

```python
%%time

# Declare regex.
empty_note = re.compile(br'{{NOTE}}[\s]*$')
full_note = re.compile(br'{{NOTE}}[\s]*\w+')
file_regex = re.compile(r'.*/(.*).docx')

for file in list_of_files:
    reFile = str(file_regex.search(file).group(1))
    txt = textract.process(file)
    
    with open(abs_dir + 'Data/Output/CleanedDocs/RBT_WET_08-2021/' + reFile + '.txt', 'bw+') as output:
    
    #     If file has an empty NOTE marker, remove NOTE marker before writing.
        if empty_note.search(txt):
            output.write(empty_note.sub(br'', txt))

    #     If file has a full NOTE, write as-is.
        elif full_note.search(txt):
            output.write(txt)

    #     If neither full nor empty, write as-is but print file name for inspection.
        else:
            print (file)
```
